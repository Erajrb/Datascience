#### <font color = Red>8. Perform filter operation by filtering people with salaries greater than 100000 USD

```python
# create a variable where we check and store the data of all the rows with salary greater than 100000USD as panda series

filt_salary = df.salary_in_usd > 100000
filt_salary
```

```python
# displaying the panda series as a df by accessing all the data from df using .loc while passing the condition as argument

df.loc[filt_salary]
```

#### <font color = Red>10. Perform double filtration: Based on job_title “Data Scientist” and residence “US”. Retrieve the list of their salaries

```python
# again create a variable to check and store the conditions

double_filtration = (df['job_title'] == 'Data Scientist') & (df['employee_residence'] == 'US')
double_filtration
```

```python
#use value_counts() fxn to count the number of true and false checked from our condition

double_filtration.value_counts()
```

```python
# use .loc and .head to display the first 5 elements of the dataframe that meet the double filter conditions

df.loc[double_filtration].head()
```

```python
#use .loc to only access the salaries in the double fitered dataframe 

df.loc[double_filtration].salary
```
